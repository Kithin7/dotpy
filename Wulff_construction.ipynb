```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#input
step = 3  #quality of graph in deg; recommended 1,2,3,5
theta = np.arange(0,360+step,step)  #in deg
theta = theta*np.pi/180  #convert deg to rad
###########################################################################
#EQUATION HERE     
#gamma = abs(np.cos(3*theta)) + abs(np.sin(3*theta)) #12-sides
gamma = abs(np.cos(3/2*theta)) + abs(np.sin(3/2*theta)) #hexagonal
#gamma = abs(np.cos(theta)) + abs(np.sin(theta))   #cubic
############################################################################
#convert polar to cartesian 
x=gamma*np.cos(theta)
y=gamma*np.sin(theta)
```

```python
#add perpendicular lines (dot) intersections
#for loop to do each      
pLineX = np.arange(2*min(x),2*max(x),max(x)/50)
for k in range(0,len(theta)):
    if y[k] == 0 or x[k] == 0: #need to handle 0,inf case
        continue #if slope = 0 or inf then skip loop iteration
    else:
        slope = (y[k]) / (x[k])
        pslope = -1/slope 
        pLineY = pslope*(pLineX-(x[k]))+(y[k])
        plt.plot(pLineX,pLineY,'b')
plt.plot(x,y,'r',linewidth=4)
plt.plot(0,0,'g.')
plt.xlim([1.2*min(x),1.2*max(x)])
plt.ylim([1.2*min(y),1.2*max(y)])
```
